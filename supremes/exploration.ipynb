```python
def get_cases_for_term(term: int):
    url = f"https://api.oyez.org/cases?per_page=0&filter=term:{term}"
    docket = helpers.load_from_remote(url)
    cases = [Case.from_id(case['term'], case['docket_number']) for case in docket]
    return cases
```
